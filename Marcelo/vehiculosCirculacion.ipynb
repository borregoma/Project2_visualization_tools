### Nota
Este código importa todos los archivos .csv que esten en la carpeta VehiculosExportados y los agrupa en un solo DF, contiene los datos de Vehículos Exportados del 2005-2021. 

```python
import pandas as pd
import glob
import geojson
import geopandas as gpd

path = "VehiculosCirculacion" # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
statesTable = pd.read_csv("tc_entidad.csv")
statesDF = pd.merge(frame, statesTable, how="left", on=["ID_ENTIDAD", "ID_ENTIDAD"])
```

```python
statesDF["TOTAL_AUTOS"] = statesDF.AUTO_OFICIAL + statesDF.AUTO_PUBLICO + statesDF.AUTO_PARTICULAR
statesDF["TOTAL_CAMIONES"] = statesDF.CAM_PAS_OFICIAL + statesDF.CAM_PAS_PUBLICO + statesDF.CAM_PAS_PARTICULAR + statesDF.CYC_CARGA_OFICIAL + statesDF.CYC_CARGA_PUBLICO + statesDF.CYC_CARGA_PARTICULAR
statesDF["TOTAL_MOTOS"] = statesDF.MOTO_OFICIAL + statesDF.MOTO_DE_ALQUILER + statesDF.MOTO_PARTICULAR
statesDF["TOTAL_GENERAL"] = statesDF.TOTAL_AUTOS + statesDF.TOTAL_CAMIONES + statesDF.TOTAL_MOTOS
statesDF.columns
```

```python
statesClean = statesDF[['ANIO', 'NOM_ENTIDAD', 'TOTAL_AUTOS', 'TOTAL_CAMIONES', 'TOTAL_MOTOS', 'TOTAL_GENERAL']]
statesCleanGrouped = statesClean.groupby(["NOM_ENTIDAD", "ANIO"])["TOTAL_GENERAL"].sum().unstack().reset_index()
statesTotalMatrix = statesCleanGrouped.fillna(0).rename_axis("", axis="columns").rename(columns={
    "NOM_ENTIDAD":"State",
    1980:"total1980",
    1981:"total1981",
    1982:"total1982",
    1983:"total1983",
    1984:"total1984",
    1985:"total1985",
    1986:"total1986",
    1987:"total1987",
    1988:"total1988",
    1989:"total1989",
    1990:"total1990",
    1991:"total1991",
    1992:"total1992",
    1993:"total1993",
    1994:"total1994",
    1995:"total1995",
    1996:"total1996",
    1997:"total1997",
    1998:"total1998",
    1999:"total1999",
    2000:"total2000",
    2001:"total2001",
    2002:"total2002",
    2003:"total2003",
    2004:"total2004",
    2005:"total2005",
    2006:"total2006",
    2007:"total2007",
    2008:"total2008",
    2009:"total2009",
    2010:"total2010",
    2011:"total2011",
    2012:"total2012",
    2013:"total2013",
    2014:"total2014",
    2015:"total2015",
    2016:"total2016",
    2017:"total2017",
    2018:"total2018",
    2019:"total2019"
})
statesTotalMatrix
```
